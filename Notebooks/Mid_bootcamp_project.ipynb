### Transforming columns into integers:

However, on the dataframe we are seeing discrete values since none of them can be measured by splitting. So we'll be transforming values in these columns into integers.

### Finding out the winner at half time:

For this we'll create a new column with the 'winner at half time' through the following function:
 - If the home team has more goals than the away team at half time, return the home team name.
 - If the home team has less goals than the away team at half time, return the away team name.
 - For every other result (home team has the same goals than the away team at half time) return 'Tie'.

```python
def half_time(row):
    
    if row['half-time_home_goals'] > row['half-time_away_goals']:
        return row["home_team_name"]
    elif row['half-time_home_goals'] < row['half-time_away_goals']:
        return row["away_team_name"]
    else: 
        return 'Tie'
```

```python
df['half-time_winner'] = df.apply(half_time, axis=1)
```

```python
df.head()
```

### Finding out the winner at full time:

```python
def full_time(row):
    
    if row['home_team_goals'] > row['away_team_goals']:
        return row["home_team_name"]
    elif row['home_team_goals'] < row['away_team_goals']:
        return row["away_team_name"]
    else: 
        return 'Extra Time'
```

```python
df['full-time_winner'] = df.apply(full_time, axis=1)
```

### Finding out the winner during extra time:

Now we'll remove the rows where there was no extra time and apply a function which is going to create a new column called 'extra_time_winner'. This column is going to show if there was extra time or not during the game, and if there was extra time it'll show the winner.

We will be using the split method to get the first word of the 'win_conditions' column, which luckily shows the winner of the game during extra time. The function will also have exceptions for those countries whose name are more than one word.

```python
df = df[df['win_conditions'].isna()==False]
```

```python
def get_winner_country(x):
    if (x == ' '):
        return 'No extra time'
    elif ( x.lower().split()[0] == 'win' ):
        return 'extra time'
    else:
        u = x.split(' ')[0]
        if ( u == 'Costa' ):
            return 'Costa Rica'
        elif ( u == 'Korea'):
            return 'Korea DPR'
        elif ( u == 'Northen'):
            return 'Northen Ireland'
        elif ( u == 'Soviet'):
            return 'Soviet Union'
        elif ( u == 'United'):
            return 'United Arab Emirates'
        elif ( u == 'Republic'):
            return 'Republic of Ireland'
        elif ( u == 'Saudi'):
            return 'Saudi Arabia'
        elif ( u == 'South'):
            return 'South Africa'
        elif ( u == 'Trinidad'):
            return 'Trinidad and Tobago'
        elif ( u == 'Serbia'):
            return 'Serbia and Montenegro'
        elif ( u == 'Czech'):
            return 'Czech Republic'
        elif ( u == 'New'):
            return 'New Zealand'
        elif ( u == 'Bosnia'):
            return 'Bosnia and Herzegovina'
        elif ( u == 'El'):
            return 'El Salvador'
        elif ( u == 'Dutch'):
            return 'Dutch East Indies'
        else:
            return u

df['extra_time_winner'] = df['win_conditions'].apply(get_winner_country)
```
