### Finding out the winner at half time:

For this we'll create a new column with the 'winner at half time' through the following function:
 - If the home team has more goals than the away team at half time, return the home team name.
 - If the home team has less goals than the away team at half time, return the away team name.
 - For every other result (home team has the same goals than the away team at half time) return 'Tie'.

```python
def half_time(row):
    
    if row['half-time_home_goals'] > row['half-time_away_goals']:
        return row["home_team_name"]
    elif row['half-time_home_goals'] < row['half-time_away_goals']:
        return row["away_team_name"]
    else: 
        return 'Tie'
```

```python
df['half-time_winner'] = df.apply(half_time, axis=1)
```

```python
df.head()
```

### Finding out the winner at full time (before the penalty shootout):

We repeat the same process as before but using the home/away_team_goals. These columns show the goals during the whole game, including aggregate time but not including penalty shootouts. 

```python
def full_time(row):
    
    if row['home_team_goals'] > row['away_team_goals']:
        return row["home_team_name"]
    elif row['home_team_goals'] < row['away_team_goals']:
        return row["away_team_name"]
    else: 
        return 'Penalties'
```

```python
df['full-time_winner'] = df.apply(full_time, axis=1)
```

### Statistical hypothesis testing:  games tied at half time and penalty shootouts.

Null hypothesis: Tied games at half-time and penalty shootouts are not related.

Alternative hypothesis: Tied games at half-time and penalty shootouts are rea.

```python
winner_crosstab = df[(df['half-time_winner'] == 'Tie')|(df['full-time_winner'] == 'Penalties')]
```

```python
data_crosstab = pd.crosstab(winner_crosstab['half-time_winner'], winner_crosstab['full-time_winner'], margins = False)
data_crosstab
```

```python
st.chi2_contingency(data_crosstab)
```

```python
st.chi2.ppf(0.95,df=1519)
```

We accept the null hypothesis, there's no relation between tied games at half time and pentaly shootouts.


### Statistical hypothesis testing: how likely is for the host country to reach the final?

Null hypothesis: home team name and final stage are not related, when home team name is equal to host country.

Alternative hypothesis: host country and final are related, when home team name is equal to host country.

```python
host_crosstab = df[(df['stage'] == 'Final')|(df['home_team_name'] == df['host_country'] )]
```

```python
data_host_crosstab = pd.crosstab(host_crosstab['stage'], host_crosstab['home_team_name'], margins = False)
```

```python
st.chi2_contingency(data_host_crosstab)
```

```python
st.chi2.ppf(0.95,df=180)
```

We reject the null hypothesis, the stage and the home team are not related.
