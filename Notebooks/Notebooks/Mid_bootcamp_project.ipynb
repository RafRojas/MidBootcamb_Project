### Transforming columns into integers:

The numerical columns: 'year', 'home_team_goals', 'attendance', 'away_team_goals', 'half-time_home_goals' and 'half-time_away_goals' are all floats (continuous) values. However on the dataframe we are seeing discrete values since none of them can be split. So we'll be transforming values in these columns into integers.

```python
df['year'] = df['year'].apply(np.int64)
df['home_team_goals'] = df['home_team_goals'].apply(np.int64)
df['away_team_goals'] = df['away_team_goals'].apply(np.int64)
df['half-time_home_goals'] = df['half-time_home_goals'].apply(np.int64)
df['half-time_away_goals'] = df['half-time_away_goals'].apply(np.int64)

df.dtypes
```

```python
def half_time(df):
    df['half-time_winner'] = []
    if ['half-time_home_goals'] > ['half-time_away_goals']:
        ['half-time_winner'] == ['home_team_name']
    elif ['half-time_home_goals'] < ['half-time_away_goals']:
        ['half-time_winner'] == ['away_team_name']
    else: 
        ['half-time_winner'] == 'Tie'
```

```python
df['half-time_winner'] = df.apply(half_time)
```
